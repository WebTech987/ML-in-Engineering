```python
# Word Frequency Analysis
def word_frequency_analysis(text_column):
    all_text = ' '.join(text_column).lower()
    word_tokens = word_tokenize(all_text)
    words_filtered = [word for word in word_tokens if word.isalpha() and word not in stopwords.words('english')]
    word_freq = nltk.FreqDist(words_filtered)
    return word_freq

# Plot Word Cloud
def plot_word_cloud(word_freq):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Product Category Distribution
def plot_category_distribution(category_column):
    plt.figure(figsize=(10, 5))
    category_column.value_counts().plot(kind='bar')
    plt.xlabel('Product Category')
    plt.ylabel('Count')
    plt.title('Product Category Distribution')
    plt.show()


# Perform analysis on the 'Product Description' column
description_column = data['Product Name']

# Word Frequency Analysis and Word Cloud
print("Combined Products")
word_freq = word_frequency_analysis(description_column)
plot_word_cloud(word_freq)

# Product Category Distribution
"""category_column = data['Product Name']
plot_category_distribution(category_column)
"""

# Perform analysis on the 'Product Description' column
description_column = data['Grassroot/Advanced']

print("Grassroots and advanced classification")
# Word Frequency Analysis and Word Cloud
word_freq = word_frequency_analysis(description_column)
plot_word_cloud(word_freq)

# Product Category Distribution
category_column = data['Grassroot/Advanced']
plot_category_distribution(category_column)
```
